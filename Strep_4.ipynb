```python
import pandas as pd
import re

data = pd.read_csv('/home/vophilong/Documents/Deep_learning/Dataset/dataset.csv')
display(data.head())
```

```python
# Chọn đặc trưng 
columns_to_keep = ['Context', 'EmailSend', 'EmailReply']
df_prep = data[columns_to_keep].copy()
print(f"[1] Kích thước ban đầu: {df_prep.shape[0]} dòng.")
# in ra 30 dòng đầu tiên để kiểm tra
display(df_prep.head(30))
```

```python
# Xử lý missing value
df_prep['Context'] = df_prep['Context'].fillna("") # Điền giá trị rỗng cho các dòng bị thiếu trong cột 'Context'
# Xóa các dòng có giá trị thiếu trong cột 'EmailSend' hoặc 'EmailReply'
df_prep = df_prep.dropna(subset=['EmailSend', 'EmailReply'])
print(f" Kích thước sau khi xử lý missing value: {df_prep.shape[0]} dòng.")
display(df_prep.head(30))
```

```python
#Biến đổi và làm sạch dữ liệu
def clean_email(text):
    if pd.isna(text):
        return ""
    # Loại bỏ các ký tự đặc biệt và số, chỉ giữ lại chữ cái và khoảng trắng
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Chuyển về chữ thường
    text = text.lower()
    # Chuẩn hóa inconsistent formatting (ví dụ: "Hello   World" -> "Hello World")
    text = re.sub(r'\s+', ' ', text).strip()
    #Loại bỏ các xuống dòng và tab
    text = re.sub(r'[\n\t]+', ' ', text)
    # loại bỏ các ký tự không in được
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    # loại bỏ các ký tự lặp lại quá nhiều lần (ví dụ: "hiiiiii" -> "hii")
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    return text
# Áp dụng cho cả 3 cột
for col in ['Context', 'EmailSend', 'EmailReply']:
    df_prep[col] = df_prep[col].apply(clean_email)
# Xem kích thước sau khi làm sạch và 30 dòng đầu tiên để kiểm tra
print(f" Kích thước sau khi làm sạch dữ liệu: {df_prep.shape[0]} dòng.")

display(df_prep.head(50))
```

```python
#Loại bỏ nhiễu


# Xử lý Skewness và Outliers bằng cách cắt tỉa độ dài của EmailSend và EmailReply
mask = (
    (df_prep['EmailSend'].str.len() >= 10) & 
    (df_prep['EmailSend'].str.len() <= 1500) & 
    (df_prep['EmailReply'].str.len() >= 10) & 
    (df_prep['EmailReply'].str.len() <= 1500)
)

df_clean = df_prep[mask].copy()# Tạo một DataFrame mới chỉ chứa các dòng hợp lệ
dropped_rows = df_prep.shape[0] - df_clean.shape[0]# Tính số dòng đã bị loại bỏ
print(f"Đã cắt tỉa {dropped_rows} dòng dữ liệu nhiễu (Outliers).")
print(f"    Kích thước sau khi lọc: {df_clean.shape[0]} dòng.")
display (df_clean.head(30))
```

```python
# Mã hóa đặc trưng
df_clean['Input_Text'] = "context: " + df_clean['Context'] + " | email: " + df_clean['EmailSend'] # Tạo cột mới 'Input_Text' bằng cách kết hợp 'Context' và 'EmailSend'
df_clean['Output_Text'] = df_clean['EmailReply']# Tạo cột mới 'Output_Text' bằng cách sao chép nội dung từ 'EmailReply'
# Chọn lại các cột cần thiết
final_dataset = df_clean[['Input_Text', 'Output_Text']].copy()
# Xem kích thước cuối cùng và 30 dòng đầu tiên để kiểm tra
print(f" Kích thước cuối cùng sau khi mã hóa và chuẩn hóa: {final_dataset.shape[0]} dòng.")
pd.set_option('display.max_colwidth', 120) # Mở rộng cột để dễ đọc
display(final_dataset.head(30))
```

```python
# Lưu dataset đã làm sạch và mã hóa vào file CSV mới
final_dataset.to_csv('/home/vophilong/Documents/Deep_learning/DoAn_Email/final_dataset.csv', index=False)
print("Đã lưu dataset đã làm sạch và mã hóa vào file 'final_dataset.csv'.")
```
