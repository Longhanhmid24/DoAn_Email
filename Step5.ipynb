Chia tập train/test

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import T5Tokenizer
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
data = pd.read_csv('/home/vophilong/Documents/Deep_learning/DoAn_Email/dataset/final_dataset.csv')
# Xem 5 dòng đầu tiên của dữ liệu
display (data.head())
```

```python
# Chia dữ liệu theo tỷ lệ chuẩn: 80% Train, 10% Validation, 10% Test
train_df, temp_df = train_test_split(data, test_size=0.2, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

print(f"[1] Phân chia tập dữ liệu thành công:")
print(f"    - Tập Huấn luyện (Train): {len(train_df)} dòng")
print(f"    - Tập Xác thực (Val):   {len(val_df)} dòng")
print(f"    - Tập Kiểm tra (Test):  {len(test_df)} dòng")

# Lưu các tập dữ liệu đã chia ra thành các file CSV riêng biệt
train_df.to_csv('/home/vophilong/Documents/Deep_learning/DoAn_Email/dataset/dataset_split/train.csv', index=False)
val_df.to_csv('/home/vophilong/Documents/Deep_learning/DoAn_Email/dataset/dataset_split/val.csv', index=False)
test_df.to_csv('/home/vophilong/Documents/Deep_learning/DoAn_Email/dataset/dataset_split/test.csv', index=False)
```
